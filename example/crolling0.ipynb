```python
from bs4 import BeautifulSoup #웹 데이터 파싱
import requests # 서버로 html 문서를 요청
import pandas as pd #데이터 프레임, 데이터 전처리
```

```python
# 요청할 웹 주소
url = 'https://www.tripadvisor.co.kr/Restaurants-g294197-Seoul.html'
```

```python
# 서버에 html 요청
response = requests.get(url)
```

```python
response #200이면 잘 받아온거임
```

```python
response.text #
```

```python
# bs4를 사용해서 데이터 파싱
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup
```

```python
#div 태그 안에, class값이 wQjYiB7z인 값을 모두 찾아줘!
liste = soup.findAll('div',{'class':'wQjYiB7z'})
```

```python
liste
```

```python
#liste 안에 있는 값들을 개별적으로 출력해보자
for item in liste:
   #각 아이템에 있는 들어있는 값 중 텍스트만 볼래
    print(item.text)
```

```python
# liste 에 있는 각 아이템을 item에 담을께
for item in liste:
    #item 중에서 a 태그에서 주소만 뽑을래
    print('https://www.tripadvisor.co.kr'+item.find('a').get('href'))
```
