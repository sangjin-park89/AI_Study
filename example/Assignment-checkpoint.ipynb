# 4. 각 나라 별 데이터 뽑아내기
## 한국, 미국, 중국, 일본, 영국

```python
# 일본의 자료 중 누적확진자 수가 감소하는 인덱스가 0번에 위치해 [1:] 을 이용해 drop

# 한국
Korea = world_covid.loc[world_covid['나라/지역']== "South Korea"]
Korea = Korea[9:]
Korea.reset_index(drop = True,inplace = True)

# 미국
USA = world_covid.loc[world_covid['나라/지역']== "US"]
USA = USA[9:]
USA.reset_index(drop = True,inplace = True)

# 중국
China = world_covid.loc[world_covid['나라/지역']== "Mainland China"]
China = China[9:]
China.reset_index(drop = True,inplace = True)

# 일본
Japan = world_covid.loc[world_covid['나라/지역']== "Japan"]
Japan = Japan[9:]
Japan.reset_index(drop = True, inplace = True)

# 영국
UK = world_covid.loc[world_covid['나라/지역']== "UK"]
UK.reset_index(drop = True, inplace = True)
```

```python
# 나라 별 일별 확진자 수 column 추가해주기
Korea['일별 확진자'] = np.zeros(len(Korea['확진자 수'].tolist()))
UK['일별 확진자'] = np.zeros(len(Korea['확진자 수'].tolist()))
Japan['일별 확진자'] = np.zeros(len(Korea['확진자 수'].tolist()))
USA['일별 확진자'] = np.zeros(len(Korea['확진자 수'].tolist()))
China['일별 확진자'] = np.zeros(len(Korea['확진자 수'].tolist()))
```

```python
# 한국
for i in range(len(Korea) - 1):
    Korea["일별 확진자"][i] = Korea['확진자 수'][i+1] - Korea['확진자 수'][i]
Korea = copy.copy(Korea[:-1])

# 미국
for i in range(len(USA) - 1):
    USA["일별 확진자"][i] = USA['확진자 수'][i+1] - USA['확진자 수'][i]
USA = copy.copy(USA[:-1])

# 중국
for i in range(len(China) - 1):
    China["일별 확진자"][i] = China['확진자 수'][i+1] - China['확진자 수'][i]
China = copy.copy(China[:-1])

# 일본
for i in range(len(Japan) - 1):
    Japan["일별 확진자"][i] = Japan['확진자 수'][i+1] - Japan['확진자 수'][i]
Japan = copy.copy(Japan[:-1])

# 영국
for i in range(len(UK) - 1):
    UK["일별 확진자"][i] = UK['확진자 수'][i+1] - UK['확진자 수'][i]
UK = copy.copy(UK[:-1])
```

```python
# 영국 이상치 데이터 수정
for i in range(len(UK['일별 확진자'].tolist())):
    if(UK['일별 확진자'][i] < 0):
        UK['일별 확진자'][i] = 0
        
# 일본 이상치 데이터 수정
for i in range(len(Japan['일별 확진자'].tolist())):
    if(Japan['일별 확진자'][i] < 0):
        Japan['일별 확진자'][i] = 0
```

# 나라 별 사망률

### 사망률 = 누적 사망자 수 / 누적 확진자 수 x 100

```python
temp = world_covid.loc[world_covid['관찰일자'] == "2021/05/29"]

temp['사망률'] = temp['사망자 수'] / temp['확진자 수'] * 100
```

### 확진자 수가 1000명 이하인 나라의 경우, 
### 사망률을 구하기 위한 모집단의 크기가 작아 공신력이 떨어질 수 있다고 생각하여 drop

```python
indexNames = temp[temp['확진자 수'] <= 1000 ].index

temp.drop(indexNames , inplace=True)
```

```python
temp['gdp 순위'] = 300
temp.reset_index(drop = True, inplace = True)
```

```python
gdp = pd.read_csv("./data/GDP.csv",header = None)

del gdp[1]
del gdp[3]
del gdp[4]

gdp.dropna(inplace = True)
gdp = gdp[:216]
```

```python
gdp.head()
```

```python
for i in range(len(gdp[2])):
    for j in range(len(temp['나라/지역'])):
        if(gdp[2][i] == temp['나라/지역'][j]):
            temp['gdp 순위'][j] = i
```

```python
indexNames = temp[temp['gdp 순위'] == 300 ].index

temp.drop(indexNames , inplace=True)
```

```python
temp.sort_values(by = ['gdp 순위'], inplace = True)
temp.reset_index(drop = True,inplace = True)
```

```python
plt.figure(figsize=(10, 10))

ax = plt.scatter(x = temp['gdp 순위'], y = temp['사망률'], alpha=0.5)
plt.title('GDP 순위 별 사망률', loc='right', pad=20)
```
